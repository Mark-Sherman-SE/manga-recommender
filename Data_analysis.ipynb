```python
for elem in ["genres", "themes", "demographics"]:
    df[elem] = df[elem].apply(lambda x: x.replace('\n','')
                                .replace('[','')
                                .replace(']','')
                                .replace('\"','')
                                .replace('\'','')
                                .replace('  ',' ')
                                .replace(', ',',')
                                .split(',')
```

```python
genres_unique = list(set(itertools.chain.from_iterable(df["genres"])))
themes_unique = list(set(itertools.chain.from_iterable(df["themes"])))
demographics_unique = list(set(itertools.chain.from_iterable(df["demographics"])))
```
